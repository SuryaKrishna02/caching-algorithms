```python
from tqdm import tqdm
import random
import sys
sys.path.append("../")
import numpy as np
np.random.seed(42)
import matplotlib.pyplot as plt

from cachingalgo.full_observation.single_cache import LFULite, LRU
from cachingalgo.full_observation.multiple_cache import LRUm, fLRU
from cachingalgo.request_generation.continuous import szipf
```

### Hit Rate Analysis for various Cache Size of  2-LRU, LRU and LRU(10)[f=2, v=1] (Static Zipf)

```python
# Request Generation
szdata = szipf(count=1_00_000, a=1, L=1_61_085) # Static Zipf Data

# Defining Parameters
L = szdata['prob'].shape[0]
req = szdata['req']
totalreq = req.shape[0]
cache = (np.array([10, 50, 100, 500, 1000, 2500, 5000, 7500, 10000], dtype='int')/2).astype(int)
Hit_rate_1 = [0]*len(cache)
Hit_rate_2 = [0]*len(cache)
Hit_rate_3 = [0]*len(cache)

# LRU(m) [f=2,v=1]
for itr, cache_size in enumerate(tqdm(cache)):
    hits1 = 0
    hits2 = 0
    hits3 = 0
    
    alg1 = LRUm(size = cache_size, f = 2, v = 1, L = L)
    alg2 = LRU(cache_size = 2*cache_size, L = L)
    alg3 = fLRU(size = cache_size, f = 2, L = L)
    
    for i in range(totalreq):
        request = req[i]

        if request in alg1:
            hits1 += 1
        
        if request in alg2:
            hits2 += 1
        
        if request in alg3:
            hits3 += 1

        alg1.update(request)
        alg2.update(request)
        alg3.update(request)
        
    Hit_rate_1[itr] = hits1/totalreq
    Hit_rate_2[itr] = hits2/totalreq
    Hit_rate_3[itr] = hits3/totalreq
    
plt.plot(cache, Hit_rate_1, '-bs', label = 'LRU(m) [f=2,v=1]', linewidth=2)
plt.plot(cache, Hit_rate_2, '-g^', label = 'LRU', linewidth=2)
plt.plot(cache, Hit_rate_3, '-r.', label = '2-LRU', linewidth=2)

plt.xlabel('Cache Size (m)')
plt.ylabel('Hit Rate')
plt.legend()
plt.savefig('../plots/full_observation/multiple_cache/lrumfhit.eps')
plt.show() 
```

### Regret Analysis of LFU-Lite, 2-LRU, LRU and LRU(10)[f=2, v=1] (Static Zipf)

```python
# Request Generation
szdata = szipf(count=1_00_000, a=1, L=1_61_085) # Static Zipf Data

# Defining Parameters
L = szdata['prob'].shape[0]
req = szdata['req']
totalreq = req.shape[0]
cache_size = 10

optcache = np.argsort(szdata['prob'])[::-1][:2*cache_size]

# LFULite
algLL = LFULite(L=L, cache_size=2*cache_size)
RegretLL = np.zeros((totalreq,))
currcache = random.sample(range(1,L), 2*cache_size)
hitsLL = 0
RegLL = 0

for i in tqdm(range(req.shape[0])):
    request = req[i]
    
    lhs = 1 if request in optcache else 0
    if request in currcache:
        rhs = 1 
        hitsLL += 1
    else:
        rhs = 0
    
    algLL.update(request, i)
    currcache = algLL.currcache(i)

    RegLL += (lhs-rhs)
    RegretLL[i] = RegLL
    
Hit_RateLL = hitsLL/ totalreq

# LRU(10) [f=2, v=1] 
Regret2 = np.zeros((totalreq,))
alg2 = LRUm(size = cache_size, f = 2, v = 1, L = L)
hits2 = 0
Reg2 = 0

for i in tqdm(range(totalreq)):
    request = req[i]
    
    lhs = 1 if request in optcache else 0
    
    if request in alg2:
        rhs = 1
        hits2 += 1
    else:
        rhs = 0
    
    alg2.update(request)
    
    Reg2 += (lhs-rhs)
    Regret2[i] = Reg2

Hit_Rate2 = hits2/totalreq


# LRU
Regret3 = np.zeros((totalreq,))
alg3 = LRU(cache_size=2*cache_size, L=L)
hits3 = 0
Reg3 = 0

for i in tqdm(range(totalreq)):
    request = req[i]
    
    lhs = 1 if request in optcache else 0
    
    if request in alg3:
        rhs = 1
        hits3 += 1
    else:
        rhs = 0
    
    alg3.update(request)
    
    Reg3 += (lhs-rhs)
    Regret3[i] = Reg3
Hit_Rate3 = hits3/totalreq


# 2-LRU
Regret4 = np.zeros((totalreq,))
alg4 = fLRU(size=cache_size, f=2, L=L)
hits4 = 0
Reg4 = 0

for i in tqdm(range(totalreq)):
    request = req[i]
    
    lhs = 1 if request in optcache else 0
    
    if request in alg4:
        rhs = 1
        hits4 += 1
    else:
        rhs = 0
    
    alg4.update(request)
    
    Reg4 += (lhs-rhs)
    Regret4[i] = Reg4
Hit_Rate4 = hits4/totalreq

x = [i for i in range(totalreq)]
plt.plot(x,list(RegretLL),label = 'LFULite', linewidth=2)
plt.plot(x,list(Regret2), label = 'LRU(20) [f=2, v=1]', linewidth=2)
plt.plot(x,list(Regret3), label = 'LRU', linewidth=2)
plt.plot(x,list(Regret4), label = '2-LRU', linewidth=2)

plt.xlabel('Time')
plt.ylabel('Regret')
plt.legend(loc="best")
plt.savefig('../plots/full_observation/multiple_cache/litelrumfreg.eps')
plt.show() 

print("Hit Rate for LFU-Lite = ", Hit_RateLL)
print("Hit Rate for LRU(20) [f=2, v=1]", Hit_Rate2)
print("Hit Rate for LRU", Hit_Rate3)
print("Hit Rate for 2-LRU", Hit_Rate4)
```

```python
# Request Generation
szdata = szipf(count=1_00_000, a=1, L=1_61_085) # Static Zipf Data

# Defining Parameters
L = szdata['prob'].shape[0]
req = szdata['req']
totalreq = req.shape[0]
cache_size = 10

virtual = [10, 50, 100, 250, 500, 750, 1000, 1250, 2500]
Hit_rate_1 = np.zeros(len(virtual),)
Hit_rate_2 = np.zeros(len(virtual),)

# LRU(m) [f=virtual+1,v=virtual]
for itr, vir in enumerate(tqdm(virtual)):
    hits1 = 0
    hits2 = 0
    
    alg1 = LRUm(size = cache_size, f = vir+1, v = vir, L = L)
    alg2 = fLRU(size = cache_size, f = vir+1, L = L)

    for i in range(totalreq):
        request = req[i]

        if request in alg1:
            hits1 += 1
        
        if request in alg2:
            hits2 += 1
            
        alg1.update(request)
        alg2.update(request)
        
    Hit_rate_1[itr] = hits1/totalreq
    Hit_rate_2[itr] = hits2/totalreq
    
    
plt.plot(virtual, Hit_rate_1, '-bs', label = 'LRU(m)', linewidth=2)
plt.plot(virtual, Hit_rate_2, '-g^', label = 'f-LRU', linewidth=2)

plt.xlabel('No. of Virtual Caches (v)')
plt.ylabel('Hit Rate')
plt.legend()
plt.savefig('../plots/full_observation/multiple_cache/lrumfvirhit.eps')
plt.show() 
```
