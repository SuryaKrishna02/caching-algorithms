```python
from tqdm import tqdm
import random
import math
import time
import sys
sys.path.append("../")
import numpy as np
np.random.seed(42)
import matplotlib.pyplot as plt

from cachingalgo.tandem_model.dynamic import LU2
from cachingalgo.request_generation.continuous import szipf
```

## Simulations of LU2 for the Unit Time

```python
zipf = [0.5, 0.8, 1.1, 1.4, 1.7, 2.0]
l = len(zipf)
hitrate1 = np.zeros((l,))
hitrate11 = np.zeros((l,))
hitrate12 = np.zeros((l,))

hitrate2 = np.zeros((l,))
hitrate21 = np.zeros(l,)
hitrate22 = np.zeros(l,)

hitrate3 = np.zeros((l,))
hitrate31 = np.zeros(l,)
hitrate32 = np.zeros(l,)

for itr, z in enumerate(tqdm(zipf)):
    szdata = szipf(L = 400, count = 1_00_000, a = z)

    req = szdata['req']
    L = szdata['prob'].shape[0]
    totalreq = req.shape[0]
    cache_size = 10
    
    F = 500*np.ones((L,), dtype = 'int')

    # Total requests that pass through Cache 2 
    c2req1 = 0
    c2req2 = 0
    c2req3 = 0

    hits1 = 0 # Total hits
    hits11 = 0 # Hits for cache1 of algorithm 1
    hits12 = 0 # Hits for cache2 of algorithm 1
    hits2 = 0
    hits21 = 0
    hits22 = 0
    hits3 = 0
    hits31 = 0
    hits32 = 0
    
    window = round((cache_size^2)*math.log(L))
    alg1 = LU2(cache_sizes = [10, 15], L = L, F = F, arr = szdata['prob'])
    alg2 = LU2(cache_sizes = [10, 15], method = 'lfu', L = L, F = F)
    alg3 = LU2(cache_sizes = [10, 15], L = L, F = F, method = 'lfulite', window=window, freqtop = 3*cache_size, useF=True)

    for i in range(req.shape[0]):
        request = req[i]

        cachehit1 = alg1.currcache(req = request, time = i)

        alg2.update(request)
        cachehit2 = alg2.currcache(req = request, time = i)
        
        alg3.update(request, ithreq = i)
        cachehit3 = alg3.currcache(req = request, time = i, ithreq = i)
        
        
        hit1 = cachehit1['cache_hit']
        hit2 = cachehit2['cache_hit']
        hit3 = cachehit3['cache_hit']

        c2req1 += cachehit1['c2pass']
        c2req2 += cachehit2['c2pass']
        c2req3 += cachehit3['c2pass']

        hits1 += int(1 in hit1)
        hits11 += hit1[0]
        hits12 += hit1[1]

        hits2 += int(1 in hit2)
        hits21 += hit2[0]
        hits22 += hit2[1]
        
        hits3 += int(1 in hit3)
        hits31 += hit3[0]
        hits32 += hit3[1]

    hitrate1[itr] = hits1/totalreq
    hitrate11[itr] = hits11/totalreq
    hitrate12[itr] = hits12/c2req1

    hitrate2[itr] = hits2/totalreq
    hitrate21[itr] = hits21/totalreq
    hitrate22[itr] = hits22/c2req2

    hitrate3[itr] = hits3/totalreq
    hitrate31[itr] = hits31/totalreq
    hitrate32[itr] = hits32/c2req3
```

```python
x = zipf


plt.figure(figsize=(15,10))
plt.plot(x, hitrate1,  label = 'LU2 Total', linewidth=3)
plt.plot(x, hitrate11, label = 'LU2 Cache 1', linewidth=3)
plt.plot(x, hitrate12, label = 'LU2 Cache 2', linewidth=3)
plt.plot(x, hitrate2,  label = 'LU2-LFU Total', linewidth=3)
plt.plot(x, hitrate21, label = 'LU2-LFU Cache 1', linewidth=3)
plt.plot(x, hitrate22, label = 'LU2-LFU Cache 2', linewidth=3)
plt.plot(x, hitrate3,  label = 'LU2-LFULite(V)[3C] Total', linewidth=3)
plt.plot(x, hitrate31, label = 'LU2-LFULite(V)[3C] Cache 1', linewidth=3)
plt.plot(x, hitrate32, label = 'LU2-LFULite(V)[3C] Cache 2', linewidth=3)
plt.legend(loc='best',  prop={"size":15})
plt.xlabel('Zipf Parameter', fontsize=20)
plt.ylabel('Hit Rate', fontsize=20)
plt.xticks(size = 15)
plt.yticks(size = 15)
plt.savefig('../plots/tandem_model/dynamic_data/alllu2hit.eps')
plt.show()
```

## Simulations of LU2 for Poisson Request arrival

```python
zipf = [0.5, 0.8, 1.1, 1.4, 1.7, 2.0]
l = len(zipf)
hitrate1 = np.zeros((l,))
hitrate11 = np.zeros((l,))
hitrate12 = np.zeros((l,))

hitrate2 = np.zeros((l,))
hitrate21 = np.zeros(l,)
hitrate22 = np.zeros(l,)

hitrate3 = np.zeros((l,))
hitrate31 = np.zeros(l,)
hitrate32 = np.zeros(l,)

arr_times = np.random.exponential(scale = 0.5, size = 10_000) # Arrival Times
for itr, z in enumerate(tqdm(zipf)):
    szdata = szipf(L = 400, count = 10_000, a = z)

    req = szdata['req']
    L = szdata['prob'].shape[0]
    totalreq = req.shape[0]
    cache_size = 10
    
    F = 500*np.ones((L,), dtype = 'int')
    req_arr_times = zip(szdata['req'], arr_times)

    # Total requests that pass through Cache 2 
    c2req1 = 0
    c2req2 = 0
    c2req3 = 0

    hits1 = 0 # Total hits
    hits11 = 0 # Hits for cache1 of algorithm 1
    hits12 = 0 # Hits for cache2 of algorithm 1
    hits2 = 0
    hits21 = 0
    hits22 = 0
    hits3 = 0
    hits31 = 0
    hits32 = 0
    
    window = round((cache_size^2)*math.log(L))
    alg1 = LU2(cache_sizes = [10, 15], L = L, F = F, arr = szdata['prob'])
    alg2 = LU2(cache_sizes = [10, 15], method = 'lfu', L = L, F = F)
    alg3 = LU2(cache_sizes = [10, 15], L = L, F = F, method = 'lfulite', window=window, freqtop = 3*cache_size, useF=True)

    sim_start = time.time()
    upd_time = 0
    ithreq = 0

    for request, delay in req_arr_times:
        # Checking whether delay time is positive or not
        if delay-upd_time >= 0:
            # Delaying the program in order mimic the poisson process
            time.sleep(delay-upd_time)

        upd_start = time.time()

        cachehit1 = alg1.currcache(req = request, time = time.time()-sim_start)

        alg2.update(request)
        cachehit2 = alg2.currcache(req = request, time = time.time()-sim_start)
        
        alg3.update(request, ithreq = ithreq)
        cachehit3 = alg3.currcache(req = request, time = time.time()-sim_start, ithreq = ithreq)
        
        
        hit1 = cachehit1['cache_hit']
        hit2 = cachehit2['cache_hit']
        hit3 = cachehit3['cache_hit']

        c2req1 += cachehit1['c2pass']
        c2req2 += cachehit2['c2pass']
        c2req3 += cachehit3['c2pass']

        hits1 += int(1 in hit1)
        hits11 += hit1[0]
        hits12 += hit1[1]

        hits2 += int(1 in hit2)
        hits21 += hit2[0]
        hits22 += hit2[1]
        
        hits3 += int(1 in hit3)
        hits31 += hit3[0]
        hits32 += hit3[1]
        
        # Calculating the time taken by the algorithm to update
        upd_time = time.time() - upd_start  

        # Increasing the request counter
        ithreq += 1

    hitrate1[itr] = hits1/totalreq
    hitrate11[itr] = hits11/totalreq
    hitrate12[itr] = hits12/c2req1

    hitrate2[itr] = hits2/totalreq
    hitrate21[itr] = hits21/totalreq
    hitrate22[itr] = hits22/c2req2

    hitrate3[itr] = hits3/totalreq
    hitrate31[itr] = hits31/totalreq
    hitrate32[itr] = hits32/c2req3
```

```python
x = zipf


plt.figure(figsize=(15,10))
plt.plot(x, hitrate1,  label = 'LU2 Total', linewidth=3)
plt.plot(x, hitrate11, label = 'LU2 Cache 1', linewidth=3)
plt.plot(x, hitrate12, label = 'LU2 Cache 2', linewidth=3)
plt.plot(x, hitrate2,  label = 'LU2-LFU Total', linewidth=3)
plt.plot(x, hitrate21, label = 'LU2-LFU Cache 1', linewidth=3)
plt.plot(x, hitrate22, label = 'LU2-LFU Cache 2', linewidth=3)
plt.plot(x, hitrate3,  label = 'LU2-LFULite(V)[3C] Total', linewidth=3)
plt.plot(x, hitrate31, label = 'LU2-LFULite(V)[3C] Cache 1', linewidth=3)
plt.plot(x, hitrate32, label = 'LU2-LFULite(V)[3C] Cache 2', linewidth=3)
plt.legend(loc='best',  prop={"size":15})
plt.xlabel('Zipf Parameter', fontsize=20)
plt.ylabel('Hit Rate', fontsize=20)
plt.xticks(size = 15)
plt.yticks(size = 15)
plt.savefig('../plots/tandem_model/dynamic_data/poissonalllu2hit0.5.eps')
plt.show()
```
